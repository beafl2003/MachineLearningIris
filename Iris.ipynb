```python
import pandas as pd
base = pd.read_csv("/content/iris.data")

base

previsores = base.iloc[: , :4].values
previsores
classe = base.iloc[:,4].values

classe 
```

```python
from sklearn.preprocessing import LabelEncoder

labelencoder_classe = LabelEncoder()
classe = labelencoder_classe.fit_transform(classe)

classe
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

previsores = scaler.fit_transform(previsores)

previsores
```

```python
from sklearn.model_selection import train_test_split

previsores_treinamento, previsores_teste, classe_treinamento, class_test = train_test_split(previsores, classe, test_size=0.25, random_state=0)

```

```python
from sklearn.naive_bayes import GaussianNB
classificador = GaussianNB()
classificador.fit(previsores_treinamento, classe_treinamento)
previsoes = classificador.predict(previsores_teste)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
acuracia = accuracy_score(class_test, previsoes)
acuracia
```

```python
matriz = confusion_matrix(class_test, previsoes) 
matriz
```
